```python
df = pd.read_csv('tips.csv')
df
```

```python
def last_digits(num):
    return str(num)[-4::]
```

```python
df['last_4numbers'] = df['CC Number'].apply(last_digits)
df
```

```python
df.info()
```

```python
def price_check(num):
    if num <= 10:
        return 'Low'
    elif num <= 15:
        return 'Meduim'
    else:
        return 'High'
```

```python
df['price_range'] = df['price_per_person'].apply(price_check)
df
```

```python
high_price_range = df[df['price_range'].isin(['High'])] ## check only isin() for practice
```

# apply method for multiple columns 
also using with lambda fucntion 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('tips.csv')
```

```python
df
```

```python
def quality (total_bill, tip):
    if total_bill/tip > 0.75:
        return "Excellent"
    else:
        return 'other'
        
```

```python
df['quality'] = np.vectorize(quality)(df['total_bill'],df['tip'])
```

```python
df
```

```python
df = df.drop( columns=['ratio of tip '],axis=1)
```

```python
df
```

```python
df['quality'] = df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'],df['tip']), axis =1)
```
