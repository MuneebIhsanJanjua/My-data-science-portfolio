# Distribution Plots

## there are three  Distribution plots types 
1) rug plot
2) histogram plot
3) kde plot

## rug plot 

```python
df.columns
```

```python
sns.rugplot(data = df , x = 'salary', height = 0.5)
plt.show()
```

```python
sns.set(style= 'darkgrid')
sns.displot(data=df , x = 'salary' ,bins = 100, kde= True)
```

# Categorical distribution

## There are two main categorical distributions for categorical data:

1. Count plot
2. Bar plot 

### count plot

```python
df = pd.read_csv('/home/mebi/Downloads/UNZIP_FOR_NOTEBOOKS_FINAL (1)/05-Seaborn/dm_office_sales.csv')
```

```python
df
```

```python
df['division'].unique()
```

```python
df['level of education'].unique()
```

```python
plt.figure(figsize=(10,4), dpi =200)
sns.set_style("dark")
sns.countplot(data= df ,x = 'work experience',palette='Set1', hue ='level of education')
```

```python
df.columns
```

```python
plt.figure(figsize=(10,8))
sns.barplot(data=df, y='work experience',x ='level of education',estimator=np.mean, hue='division')
plt.legend(bbox_to_anchor= (1,1))
```
